```python
import matplotlib.pyplot as plt
```

```python
# Set the X axis to a list of strings for each month

x_axis = ["Jan","Feb","Mar","Apr","May","June","July","Aug","Sept","Oct","Nov","Dec"]

# Set y-axis to a list of floats as the total fare in US dollars accumlated for each month

y_axis = [10.02, 23.24, 39.20, 35.42, 32.42, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create a plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()

ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()

fig = plt.figure()

ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
#create a quick plot

ax = plt.axes()

ax.plot(x_axis,y_axis)
```

```python
#create the plot and add a label for the legend
plt.plot(x_axis,y_axis, label = 'Boston')

#create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

#set the y limit between 0 and 45
plt.ylim(0,45)

#create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend() 
```

```python
#create the plot
plt.plot(x_axis, y_axis, marker="D" ,color="green", linewidth=2, label ='Boston')

#create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

#set the y limit between 0 and 45
plt.ylim(0,45)

#create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend() 

plt.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot

plt.bar(x_axis, y_axis, color = "green", label='Boston')

#create labels for the x and y

plt.xlabel("Date")
plt.ylabel("Fare($)")

#create a title

plt.title("PyBer Fare by Month")

#add a legend
plt.legend()
```

```python
# create horizontal bar graph

plt.barh(x_axis,y_axis, color = "magenta", label = 'Boston')

#invert the months

plt.gca().invert_yaxis()

#create labels for x and y axis

plt.xlabel("Fare($)")

plt.ylabel("Date")

#title

plt.title("PyBer Fare by Month")

# add a legend
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = 'Chicago')
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(y_axis,x_axis, 'd',c="red", label = 'Chicago')

plt.legend()

plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Month")

plt.xlim(0,50)

plt.gca().invert_xaxis()

plt.gca().invert_yaxis()
```

```python
#Bubble charts with markers propotionate to the data
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#make the data point bigger
y_axis_larger = []

for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
#make a pie chart
```

```python
plt.pie(y_axis, labels = x_axis)
```

```python
plt.subplots(figsize = (8,8))
# separate a piece of the pie
explode_values = (0, 0, 0, 0, 0, 0, .02, 0, 0, 0, 0, 0)

#format with percentages
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, 
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    shadow = True,
    startangle = 100,
    counterclock = False)

plt.show()
```

```python
fig, ax = plt.subplots()

explode = [0,0,0.2, 0,0,0,0.2,0,0,0,0,0]
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.figsize = (10,10)
ax.pie(y_axis,
       explode = explode,
       labels=x_axis,
       colors = colors,
       autopct = '%.1f%%',
       shadow = True,
       startangle = 100,
       counterclock = False,)
      
plt.show()
```
